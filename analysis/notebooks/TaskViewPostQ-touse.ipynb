```python
from pymongo import MongoClient
import numpy as np
import itertools

# DB connections
client = MongoClient()
db = client.db_tasklog_clean
# Collections
User = db.user
Log = db.log_chrome
UserTasks = db.user_tasks
```

```python
# Taskmap (top level only)
taskmap = {}
for t in UserTasks.find({}):
    taskmap[t['_id']] = t['task']
```

```python
# postQ stats
stats = {}
for u in User.find({}):
    if not 'postQ' in u:
        print u
        continue
    postQ = u['postQ']
    if 'questionnaire' in postQ:
        for X in postQ['questionnaire']:
            questions = postQ['questionnaire'][X]
            for q in questions:
                tmp = stats.get(q, [])
                tmp.append(questions[q])
                stats[q] = tmp
for q in stats:
    answer = stats[q]
    answer.sort()
    print q, 
    for k, g in itertools.groupby(answer):
        g = list(g)
        perc = len(g)/float(len(answer))
        print '%s (%d - %.2f)'%(k, len(g), perc), 
    print len(answer)
    print
        
```

```python
# Check each user's labels

for u in User.find({}):
    ulog = Log.find({'userid': u['userid'], 'to_annotate': True, 'removed': False})
    labels = []
    for x in ulog:
        if 'task' in x['annotation']:
            labels.append(x['annotation']['task']['taskid'])
    labels.sort()
    for k, g in itertools.groupby(labels):
        print u['userid'], k, len(list(g))
```
